```python
def f(x):
    return x**3 - x - 2       # f(x)  = x^3 - x - 2

def f_prime(x):
    return 3 * x**2 - 1        # f'(x) = 3x^2 - 1

def newton_raphson(x0, tol=0.0001, max_iter=100):
    print(f"{'Iter':<6} {'x0':<15} {'f(x0)':<15} {'f\'(x0)':<15} {'x1':<15}")
    print("-" * 66)

    for iteration in range(1, max_iter + 1):
        fx  = f(x0)
        fpx = f_prime(x0)

        if fpx == 0:
            print("Derivative is zero. Method fails.")
            return

        # Newton Raphson formula
        x1 = x0 - fx / fpx
        print(f"{iteration:<6} {x0:<15.8f} {fx:<15.8f} {fpx:<15.8f} {x1:<15.8f}")

        # Check for convergence
        if abs(x1 - x0) < tol:
            print(f"Approximate Root = {x1:.8f}")
            print(f"Total Iterations = {iteration}")
            return

        x0 = x1   # Update x0 for next iteration

    print("Method did not converge within maximum iterations.")

x0 = 1.5     # Initial guess
newton_raphson(x0)
```
