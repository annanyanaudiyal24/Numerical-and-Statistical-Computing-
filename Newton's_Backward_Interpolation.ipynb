```python
def backward_difference_table(y):
    n = len(y)
    # diff[i][j] = j-th backward difference ending at i
    diff = [[0] * n for _ in range(n)]
    for i in range(n):
        diff[i][0] = y[i]
    for j in range(1, n):
        for i in range(j, n):
            diff[i][j] = diff[i][j - 1] - diff[i - 1][j - 1]
    return diff

def newton_backward(x_values, y_values, x):
    n = len(x_values)
    h = x_values[1] - x_values[0]     # Step size
    s = (x - x_values[-1]) / h          # s = (x - xn) / h

    diff = backward_difference_table(y_values)

    print("Backward Difference Table:")
    print(f"{'x':<8} {'y':<10}", end="")
    for k in range(1, n): print(f"{f'∇{k}y':<10}", end="")
    print()
    print("-" * (8 + 10*n))
    for i in range(n):
        print(f"{x_values[i]:<8}", end="")
        for j in range(i + 1):
            print(f"{diff[i][j]:<10.4f}", end="")
        print()

    # Apply Newton's Backward formula
    result = diff[n - 1][0]
    s_term = 1
    factorial = 1
    for k in range(1, n):
        s_term  *= (s + (k - 1))
        factorial *= k
        result  += (s_term / factorial) * diff[n - 1][k]

    print(f"Interpolating at x = {x}")
    print(f"s = (x - xn) / h = ({x} - {x_values[-1]}) / {h} = {s:.4f}")
    print(f"f({x}) = {result:.6f}")
    return result

x_values = [1891, 1901, 1911, 1921, 1931]
y_values = [46, 66, 81, 93, 101]   # Population in millions
x = 1929   # Interpolate near the END of the table

newton_backward(x_values, y_values, x)
```
