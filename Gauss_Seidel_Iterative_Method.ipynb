```python
def gauss_seidel(A, B, tol=0.0001, max_iter=50):
    n = len(B)
    X = [0.0] * n    # Initial guess: all zeros

    print(f"{'Iter':<6}" + " ".join(f"{f'x{i+1}':<12}" for i in range(n)))
    print("-" * (6 + 12*n))

    for iteration in range(1, max_iter + 1):
        X_old = X[:]

        # Calculate each variable using UPDATED values
        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i][j] * X[j]   # Uses X (updated values)
            X[i] = (B[i] - sigma) / A[i][i]

        vals = " ".join(f"{v:<12.6f}" for v in X)
        print(f"{iteration:<6}{vals}")

        # Check convergence
        if max(abs(X[i] - X_old[i]) for i in range(n)) < tol:
            print("Converged!")
            break

    print("Solution:")
    for i in range(n):
        print(f"  x{i+1} = {X[i]:.6f}")

# Equations:
#  10x + 2y + z  = 7
#   x + 5y - 3z  = -8
#   x + y  + 8z  = 26
A = [[10, 2,  1],
     [ 1, 5, -3],
     [ 1, 1,  8]]
B = [ 7, -8, 26]

gauss_seidel(A, B)
```
