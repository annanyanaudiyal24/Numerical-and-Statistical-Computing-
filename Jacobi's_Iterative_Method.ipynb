```python
def jacobi(A, B, tol=0.0001, max_iter=50):
    n = len(B)
    X = [0.0] * n    # Initial guess: all zeros

    print(f"{'Iter':<6}" + " ".join(f"{f'x{i+1}':<12}" for i in range(n)))
    print("-" * (6 + 12*n))

    for iteration in range(1, max_iter + 1):
        X_new = [0.0] * n

        # Calculate each variable using OLD values
        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i][j] * X[j]   # Uses X (old values)
            X_new[i] = (B[i] - sigma) / A[i][i]

        vals = " ".join(f"{v:<12.6f}" for v in X_new)
        print(f"{iteration:<6}{vals}")

        # Check convergence
        if max(abs(X_new[i] - X[i]) for i in range(n)) < tol:
            print("Converged!")
            break

        X = X_new   # Update all values at once

    print("Solution:")
    for i in range(n):
        print(f"  x{i+1} = {X_new[i]:.6f}")

# Equations:
#  10x + 2y + z  = 7
#   x + 5y - 3z  = -8
#   x + y  + 8z  = 26  (diagonally dominant)
A = [[10, 2,  1],
     [ 1, 5, -3],
     [ 1, 1,  8]]
B = [ 7, -8, 26]

jacobi(A, B)
```
