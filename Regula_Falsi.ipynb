```python
def f(x):
    return x**3 - x - 2

def regula_falsi(a, b, tol=0.0001, max_iter=100):
    if f(a) * f(b) >= 0:
        print("f(a) and f(b) must have opposite signs.")
        return

    print(f"{'Iter':<6} {'a':<12} {'b':<12} {'c':<12} {'f(c)':<12}")
    print("-" * 56)

    for iteration in range(1, max_iter + 1):
        # Calculate the false position point
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        print(f"{iteration:<6} {a:<12.6f} {b:<12.6f} {c:<12.6f} {f(c):<12.6f}")

        if abs(f(c)) < tol:
            break
        elif f(a) * f(c) < 0:
            b = c   # Root is in [a, c]
        else:
            a = c   # Root is in [c, b]

    print(f"Approximate Root = {c:.6f}")
    print(f"Total Iterations = {iteration}")

a = 1
b = 2
regula_falsi(a, b)
```
