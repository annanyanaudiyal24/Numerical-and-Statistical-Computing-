```python
def f(x):
    return x**3 - x - 2

def bisection(a, b, tol=0.0001):
    if f(a) * f(b) >= 0:
        print("f(a) and f(b) must have opposite signs.")
        return

    iteration = 0
    print(f"{'Iter':<6} {'a':<12} {'b':<12} {'mid':<12} {'f(mid)':<12}")
    print("-" * 56)

    while (b - a) / 2 > tol:
        mid = (a + b) / 2
        iteration += 1
        print(f"{iteration:<6} {a:<12.6f} {b:<12.6f} {mid:<12.6f} {f(mid):<12.6f}")

        if f(mid) == 0:
            break
        elif f(a) * f(mid) < 0:
            b = mid   # Root lies in left half
        else:
            a = mid   # Root lies in right half

    root = (a + b) / 2
    print(f"Approximate Root = {root:.6f}")
    print(f"Total Iterations = {iteration}")


# Driver Code
a = 1
b = 2
bisection(a, b)
```
